# 📐 Theoretical stats

Lander-Waterman formulas for shotgun sequencing: `compute_alpha`, `expected_contigs`, `expected_contig_length`.

#### `compute_alpha`
*Compute the read density parameter α = N * L / G.*


```python
def compute_alpha(N: int, L: int, G: int) -> float:
    """Compute the read density parameter α = N * L / G.

    Args:
        N: Number of reads.
        L: Read length (maximum read length used as an approximation).
        G: Genome length.
    Returns:
        α as a float.
    """
    if G == 0:
        return 0.0
    return N * L / G
```

#### `expected_contigs`
*Expected number of contigs ≈ N · e^{-α}.*


```python
def expected_contigs(N: int, alpha: float) -> float:
    """Expected number of contigs ≈ N · e^{-α}.

    Args:
        N: Number of reads.
        alpha: Read density parameter.
    Returns:
        Expected contig count.
    """
    return N * math.exp(-alpha)
```

#### `expected_contig_length`
*Expected contig length ≈ G · (1 − e^{-α}) / (N · e^{-α}).*


```python
def expected_contig_length(G: int, N: int, alpha: float) -> float:
    """Expected contig length ≈ G · (1 − e^{-α}) / (N · e^{-α}).

    Args:
        G: Genome length.
        N: Number of reads.
        alpha: Read density parameter.
    Returns:
        Expected contig length.
    """
    if N == 0:
        return 0.0
    return G * (1 - math.exp(-alpha)) / (N * math.exp(-alpha))


# =============================================================================
# MODULE: overlap
# =============================================================================
```
